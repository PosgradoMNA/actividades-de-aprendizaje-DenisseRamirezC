ANOVA: 
-Sirve para encontrar la correlacin entre diferentes grupos de una variable categrica
-Devuelve f-test score y p-value 

F-test score: variacin entre las medias de los grupos de muestra dividida por la variacin dentro del grupo de muestra.

P-value: indica si el resultado obtenido es estadisticamente significativo.
