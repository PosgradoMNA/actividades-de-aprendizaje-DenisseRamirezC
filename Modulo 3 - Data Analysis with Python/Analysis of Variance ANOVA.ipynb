Comparacion estadistica de grupos. Identificar que valor en cada variable categórica tiene mayor impacto en la predicción del precio.

ANOVA: 
-Sirve para encontrar la correlación entre diferentes grupos de una variable categórica
-Devuelve f-test score y p-value 

F-test score: variación entre las medias de los grupos de muestra dividida por la variación dentro del grupo de muestra.

P-value: indica si el resultado obtenido es estadisticamente significativo.
