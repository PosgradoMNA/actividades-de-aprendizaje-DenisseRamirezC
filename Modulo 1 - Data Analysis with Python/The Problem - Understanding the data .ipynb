# Understanding the Data 
Datos: los datos pueden ser recolectados de forma manual por científicos o de forma digital cada vez que damos clic a algo o abrimos una página. Tomando el ejemplo del carro de Tom: quiere vender su carro pero no sabe que precio ponerle, ya que quiere un precio en el que gane dinero pero también que sea razonable para que otras personas esten interesadas en comprarlo.
