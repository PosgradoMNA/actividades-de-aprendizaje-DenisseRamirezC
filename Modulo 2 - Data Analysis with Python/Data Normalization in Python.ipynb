When two or more features are in very different ranges, one  will
intrinsically influence the result more, due to its larger value, but this doesn’t necessarily mean it is more ‘important’ as a predictor.

We can normalize these variables into values that range from 0 to 1.

METHODS OF NORMALIZING DATA 


```python
df['length'] = df['length']/df['length'].max() #simple feature scaling 
```

```python
df['length'] = (df['length'] - df['length'].min())/(df['length'].max() - df['length'].min()) #min-max
```

```python
df['length'] = (df['length']-df['length'].mean())/df['length'].std()#z-score
```
