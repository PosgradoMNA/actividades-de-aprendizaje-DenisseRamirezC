Subject: Binning in Python

Binning is grouping of values into 'bins' 
-can improve accuracy of the predictive models.
-converts numeric into categorical variables. 

for example: 

price = [1 to 4] [5 to 7] [8 to 10]
bins =   low      medium    high

```python
bins = np.linspace(min(df['price']), max(df['price']), 4)
group_names = ['low', 'medium', 'high']
df['price-bined']=pd.cut(df['price'], bins, labels = group_names, include_lowest = True)
```
